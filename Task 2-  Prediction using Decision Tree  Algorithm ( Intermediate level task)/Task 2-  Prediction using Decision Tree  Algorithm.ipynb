# 8. Making models using Decision tree classifier

```python
X = iris.drop('Species',axis=1)
Y = iris['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25)
```

```python
Decision_tree_model=DecisionTreeClassifier(max_depth=4)
Decision_tree_model.fit(x_train,y_train)
```

```python
print("Accuracy of Model: ",Decision_tree_model.score(x_train,y_train)*100)
```

# 9. Ploting Decision Tree

```python
X.columns
```

```python
fig = plt.figure(figsize=(25,20))
plot_tree(Decision_tree_model,feature_names=X.columns,  
                   class_names=list(type_species.keys()),
                   filled=True)
```
