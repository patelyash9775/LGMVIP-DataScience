# **Task-4 - Image to Pencil Sketch with Python**


## **1. Importing the all required Libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
```

## **2. Reading and Displaying the Image**

```python
flower_image=cv2.imread("/flower.jpeg")
cv2_imshow(flower_image)
```

## **3.converting the image into grey scale image.**

```python
gray_image=cv2.cvtColor(flower_image,cv2.COLOR_BGR2GRAY) 
cv2_imshow(gray_image)
```

## **4. Inverting the gray scale image**

```python
inverted_img = cv2.bitwise_not(gray_image)  
cv2_imshow(inverted_img)
```

## **5. Using the Gaussian blur method for smoothing the image**

```python
blurred_img = cv2.GaussianBlur(inverted_img, (21, 21), 0)
cv2_imshow(blurred_img)
```

# **6. Making the final pencil sketch**

```python
Inverted_Blurred_flower_img = 255 - blurred_img
pencil_sketch_image = cv2.divide(gray_image, Inverted_Blurred_flower_img, scale=256)
cv2_imshow(pencil_sketch_image)
```
